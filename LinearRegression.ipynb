**HOME RESNTALPREDEDICTION**

```python
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df=pd.read_csv('/content/home-rental.csv')

x=df[['size']]
y=df[['rent_amount']]
```

```python
from numpy.random.mtrand import random
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
print("mean squared error:",mse)
```

```python
plt.scatter(x,y,color="blue")
plt.plot(x,model.predict(x),color='red')
plt.xlabel('size')
plt.ylabel('rent_amount($1000)')
plt.title('home rental analysis')
plt.show()

```

MALL CUSTEMER SEGMENTATION

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
import pandas as pd
df=pd.read_csv('/content/Mall_Customers.csv')
```

```python
x=df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
agg_model=AgglomerativeClustering(n_clusters=3)
```

```python
df['cluster']=agg_model.fit_predict(x)
```

```python
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'],c=df['cluster'],cmap='rainbow')
plt.xlabel('annual income')
plt.ylabel('spending score')
plt.title('Aggleomerative clustering example')
plt.show()
```

Home rental anlysis(Kmeans)


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('/content/home-rental.csv')
```

```python
kmeans=KMeans(n_clusters=3,random_state=42)
X = df[['size', 'bedrooms', 'rent_amount']]
df['cluster']=kmeans.fit_predict(X)
```

```python
controids=kmeans.cluster_centers_
```

```python
plt.scatter(df['size'], df['rent_amount'], c=df['cluster'], cmap='rainbow')
plt.scatter(controids[:,0], controids[:,2], color='black', marker='x', s=100)
plt.xlabel('size')
plt.ylabel('rent_amount')
plt.title("home rental analysis")
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt
from google.colab import files
```

```python
image_file='/content/OIP (5).webp'
image=cv2.imread(image_file)
```

```python
!wget -O haar_face.xml https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml
face_cascade=cv2.CascadeClassifier('haar_face.xml')
```

```python
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
faces=face_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=5)
```

```python
for(x,y,w,h)in faces:
    cv2.rectangle(image,(x,y),(x+w, y+h),(255,0,0),2)
```

```python
plt.figure(figsize=(8,6))
plt.imshow(image)
plt.title(f'Faces detected:{len(faces)}')
plt.axis('off')
plt.show()
```
